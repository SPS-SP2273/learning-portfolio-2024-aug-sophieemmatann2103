### Discuss with a friend to understand what each of the numbers in the %%timeit output represents and write down their meanings.

38.6 ns shows the average time it took to execute the code in one loop. 

 0.664 is the standard deviation, which shows how much the execution times varied across the different runs. A smaller value will indicate that the execution time was consistent across the runs.

7 runs means that the %%timeit command ran the entire set of loops 7 times to calculate the average execution time and standard deviation. 

For 10,000,000 loops each, it means that in each of the 7 runs, the code was executed 10 million times. The %%timeit command determines the number of loops automatically based on how fast the code executes, aiming for a time measurement in the range of a few seconds.
