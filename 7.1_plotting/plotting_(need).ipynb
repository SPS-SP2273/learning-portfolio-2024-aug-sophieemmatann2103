```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75   # this will give the error bars which are seen as 'plus signs'

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,   # abbreviating the error names
             label='Y3 with errors', color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=0.25)  # the alpha value regulates the transparency of a graph plot
plt.legend(loc='upper left')   # shows us the labels
plt.savefig('plot.png', dpi=150)     # this will save in the notebook under plotting files
plt.show()
```

here is how we can write abbreviations (both will give the same result):

plt.plot(x, y, color='red', linestyle='dashed', linewidth=2)

plt.plot(x, y, c='red', ls='dashed', lw=2)

here is how we can use the for loop in plots and its outputs: 

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

# Lets start plotting
for alpha in [.25, .75, 1]:
    plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
    plt.plot(x, y_2, label='Y2 values', color='blue', )
    plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green')
    plt.xlabel('x-values')
    plt.ylabel('y-values')
    plt.title('X vs Y')
    plt.grid(alpha=.25)
    plt.legend(loc='upper left')
    plt.savefig('plot pdf', dpi =300)
```
